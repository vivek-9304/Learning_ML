```python
import pandas as pd
```

```python
class LinearRegression():
    def __init__(self, lr=0.01, epochs=1000):
        self.lr = lr
        self.epochs = epochs
        self. m = 0
        self. b = 0

    def predict(self, X):
        y_pred = []
        for x in X:
            y_pred.append(self.m*x + self.b)
        return y_pred

    def losses(self, Y, y_pred):
        n = len(Y)
        loss = sum((y_pred[i]-Y[i])**2 for i in range(n))/n
        return loss

    def Update(self, X, Y, y_pred):
        dm = 0
        db = 0
        n = len(X)
        for i in range(n):
            error = y_pred[i] - Y[i]
            dm += X[i]*error
            db += error
        dm = (2/n)*dm
        db = (2/n)*db
        self.m = self.m - self.lr*dm
        self.b = self.b - self.lr*db

    def fit(self, X, Y):
        for ep in range(self.epochs):
            y_pred = self.predict(X)
            self.Update(X, Y, y_pred)
            if(ep%200==0):
                print("losses be ",self.losses(Y, y_pred))
```

```python
X = [10, 11, 12, 15, 18]
Y = [5, 6, 7, 8, 10]
model = LinearRegression()
model.fit(X, Y)
```

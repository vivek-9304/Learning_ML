```python
import pandas as pd
```

```python
class LinearRegression:
    def __init__(self, lr=0.0001, epochs=1001):
        self.lr = lr
        self.epochs = epochs
        self.m = 0
        self.b = 0

    def predict(self, X):
        y_pred = []
        for x in X:
            y_pred.append(self.m*x + self.b)
        return y_pred

    def losses(self, Y, y_pred):
        n = len(Y)
        loss = sum((y_pred[i]-Y[i])**2 for i in range(n)) / n
        return loss

    def update(self, X, Y, y_pred):
        dm = 0
        db = 0
        n = len(Y)
        for i in range(n):
            error = y_pred[i]-Y[i]
            dm += X[i]*error
            db += error
        dm = (2/n)*dm
        db = (2/n)*db
        self.m = self.m - self.lr*dm
        self.b = self.b - self.lr*db

    def fit(self, X, Y):
        for ep in range(self.epochs):
            y_pred = self.predict(X)
            self.update(X, Y, y_pred)
            if(ep%200==0):
                print(f"Loss is {self.losses(Y, y_pred)}")
        return self

```

```python
X = [10, 11, 12, 15, 18]
Y = [50, 55, 60, 76, 93]
model = LinearRegression()
model.fit(X, Y)
print("Predictions:", model.predict([20]))
```
